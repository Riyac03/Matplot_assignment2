```python
def remove_outliers(df,col,k):
    mean=df[col].mean()
    global df1,var1
    sd=df[col].std()
    var1=[x for x in df[col]if (x>mean-k*sd)]
    var1=[x for x in df[col]if (x<mean+k*sd)]
    df1=df.loc[df[col].isin(var1)];
    print(df1.shape)
    print("number of outliers removed -->",df.shape[0]-df1.shape[0])
```
